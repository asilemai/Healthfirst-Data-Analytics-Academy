# Odd or Even Python Project

This program asks the user to enter a number and determines if it is odd or even.  
It also checks if the number is a multiple of 4.  
Finally, it asks for two numbers and determines if one divides evenly into the other.

```python
# ask the user for a number
num = int(input("Enter a number: "))

# check if number is even or odd
if num % 2 == 0:
    print("The number is even")
else:
    print("The number is odd")

# check if it is a multiple of 4
if num % 4 == 0:
    print("The number is also a multiple of 4")

# ask the user for two numbers
num2 = int(input("Enter a number to check: "))
check = int(input("Enter a number to divide by: "))

# check if it divides evenly
if num2 % check == 0:
    print("It divides evenly")
else:
    print("It does not divide evenly")
```
